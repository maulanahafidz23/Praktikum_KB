```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("googleplaystore.csv")
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# Membagi Dataset menjadi Training set dan Test set 

Membagi dataset menjadi training set dan test set dengan proporsi training set 70 dan test set 30

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print("Data Training : ", len(X_train))
print("Data Test : ", len(X_test))
```

# Normalisasi

Melakukan normalisasi atribut rating menggunakan MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

copy_minmax = df.copy()

copy_minmax[['Rating']] = MinMaxScaler().fit_transform(copy_minmax[['Rating']])

print(copy_minmax)
```

# Standarisasi

Melakukan standarisasi atribute rating dengan menggunakan StandardScaler

```python
from sklearn.preprocessing import StandardScaler

copy_ss = df.copy()

copy_ss[['Rating']] = StandardScaler().fit_transform(copy_ss[['Rating']])

print(copy_ss)
```

# Mengubah tipe data

Menrubah tipe data pada attribut Reviews dari object menjadi tipe data numerik

```python
print("Nilai unik dalam kolom 'Reviews : '")
print(df['Reviews'].unique())
df['Reviews'] = pd.to_numeric(df['Reviews'], errors = 'coerce')
```

```python
print("Nilai unik dalam kolom 'Reviews : '")
print(df['Reviews'].unique())
```

# Data Cleaning


```python
print("Jumlah record null : ")
print(df.isnull().sum())
```

```python
strategy = {
    'Rating' : df['Rating'].mean(),
    'Reviews' : df['Reviews'].median(),
    'Type' : df['Type'].mode()[0],
    'Content Rating' : df['Content Rating'].mode()[0],
    'Current Ver' : df['Current Ver'].mode()[0],
    'Android Ver' : df['Android Ver'].mode()[0],
}

data_cleaning =  df.fillna(value = strategy)

print("Jumlah record yang null: ")
print(data_cleaning.isnull().sum())
```

```python
print("Data yang memiliki duplikat : ")
df.duplicated().sum()
```

```python
data_cleaning.drop_duplicates(inplace = True)
data_cleaning.duplicated().sum()
print("Data yang memiliki duplikat : ", data_cleaning.duplicated().sum())
```

```python
data_cleaning.info()
```

# One Hot Encoding


```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output = False)

encoded_df = ohe.fit_transform(data_cleaning[['Type']])

# Membuat Data Frame baru dengan hasil encoding
new_df = pd.DataFrame(encoded_df, columns = ohe.get_feature_names_out(['Type']))

#Menggabungkan hasil encoding dengan dataset
data_encoded = pd.concat([data_cleaning, new_df], axis = 1)

data_encoded.head()
```
