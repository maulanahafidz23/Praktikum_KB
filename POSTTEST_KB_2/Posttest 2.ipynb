### 3.Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
# Menghitung jumlah record dan jumlah attribute
jumlah_record = len(df)
jumlah_attribute = len(df.columns)

# Mendapatkan nama attribute
nama_attribute = list(df.columns)

# Menghitung jumlah attribute dengan tipe angka
jumlah_attribute_angka = len(df.select_dtypes(include=['int64', 'float64']).columns)

# Membuat dictionary dengan informasi yang dikumpulkan
informasi_dict = {
    "Jumlah Record": jumlah_record,
    "Jumlah Attribute": jumlah_attribute,
    "Nama Attribute": nama_attribute,
    "Jumlah Attribute dengan Tipe Angka": jumlah_attribute_angka
}

# Menampilkan isi dari dictionary
for key, value in informasi_dict.items():
    print(f"{key}: {value}")
```
